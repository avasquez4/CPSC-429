## Lab 2: Data Preparation

```python
import pandas as pd
import scipy as sp
import numpy as np
```

## 0. Data Loading

```python
#load data and extract data
names = ['age', 'workclass', 'fnlwgt', 'edu', 'edu-num', 'maritalstatus', 'occupation', 'relationship', 'race', 'sex','capital-gain','capital-loss','hours-per-week','native-country','income']
df = pd.read_csv('adult.data', names=names)
print( "Total columns: ", len(df.columns))
#df.head()                 # display all columns
df[df.columns[:10]].head() # display the first 10 columns
```

## 1. Standardization 

```python
################################################
# Standardize data (0 mean, 1 stdev)
################################################
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler().fit(df[['fnlwgt']])
df[['fnlwgt']] = scalar.transform(df[['fnlwgt']])
df[df.columns[:5]].head()
```

## 2. Discretization (uniform transformation)


```python
#Import the class
from sklearn.preprocessing import KBinsDiscretizer
 
# your code below
transf = KBinsDiscretizer(n_bins = 5, encode = 'ordinal', strategy = 'uniform')

df[['age']] = transf.fit_transform(df[['age']])
df[df.columns[:5]].head()
```

## 3. One-hot encoding


```python
from sklearn.preprocessing import OneHotEncoder
 
# your code below
encoder = OneHotEncoder(sparse=False)
race_encoded = encoder.fit_transform(df[['race']])

feature_names = encoder.get_feature_names_out(input_features=['race'])
race_encoded = pd.DataFrame(race_encoded, columns=feature_names)

race_encoded.head()
```
